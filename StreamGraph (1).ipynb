```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from google.colab import files
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/data/covid_data.csv')
df
```

```python
# agrupo
agrupado =df[['date','continent','new_cases']].groupby(['date','continent']).sum()
agrupado
```

```python
#Pivotar continent
agrupado= agrupado.unstack(1)
agrupado
```

```python
# Ahora a los valores NAN ,  voy a cambiarlos  por el valor cero
agrupado = agrupado.fillna(0)
agrupado
agrupado = agrupado.droplevel(0, axis=1)
agrupado
```

```python
#crer lista
import matplotlib.pyplot as plt
values = [
    agrupado['Africa'],
    agrupado['Asia'],
    agrupado['Europe'],
    agrupado['North America'],
    agrupado['Oceania'],
    agrupado['South America']
]
```

```python
#hacer gráfico
fig, ax= plt.subplots()
fig.set_size_inches(15,10)
ax.stackplot(
    agrupado.index.values,
    values,
    baseline = 'wiggle',
    labels= agrupado.columns.values

)
ax.legend(loc='upper left')
ax.set_title('Casos de Covid por Continente')
ax.set_xlabel('Día')
ax.set_ylabel('Número de Casos')
ax.axhline(0, color="black", ls="--", linewidth=0.5)
fig.autofmt_xdate()
fig.tight_layout()
plt.xticks(np.arange(0, len(agrupado.index.values), 30))
plt.show()
```
